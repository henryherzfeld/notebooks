```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
import cv2
import time
from scipy.special import softmax
from copy import deepcopy
```

```python
MAX_TRI = 50
MIN_TRI = 20
tri_mutation_size = 50
IMG_BASE = np.full((225, 225, 3), 255, dtype=np.uint8)
BEST_FIT = deepcopy(IMG_BASE)

target = cv2.imread("smiley.png")

class Triangle:
    def __init__(self, xyz, rgb):
        self.xyz = xyz
        self.rgb = rgb
        
    def fitness(self):
        img_base = deepcopy(IMG_BASE)
        cv2.fillPoly(img_base, self.xyz, self.rgb, cv2.LINE_AA)
        
        cv2.addWeighted(img_base, 0.5, BEST_FIT, 0.5, 0) #alpha blending
        
        diff = sum(np.abs(img_base - target))
        target_tot = sum(target)
        return (target_tot - diff) / target_tot
    
    def mutate(self, resolution):
        # mutate coords
        new_xyz = []
        rands = np.random.randint(-tri_mutation_size, tri_mutation_size+1, self.xyz.size)
        
        for val, rand in zip(np.ndarray.flatten(self.xyz), rands):
            new_xyz.append(np.clip(val + rand, 0, resolution))
        self.xyz = np.reshape(new_xyz, self.xyz.shape)
        
        # mutate rgb
        self.rgb = tuple([int(x) for x in np.clip(self.rgb + np.random.randint(-tri_mutation_size, tri_mutation_size, 3), 0, 256)])
            

class Genes:    
    def __init__(self, resolution):
        self.resolution = resolution
        self.__n_triangles = np.random.randint(MIN_TRI, MAX_TRI+1)
        
        self.triangles = []
        for _ in range(self.n_triangles):
            self.add_triangle()
    
    def add_triangle(self):
        xyz = np.random.randint(0, self.resolution, size=(1, 3, 2))
        rgb = tuple([int(x) for x in np.random.randint(0, 256, size=3)])
        self.triangles.append(Triangle(xyz, rgb))
            
    @property
    def n_triangles(self):
        return self.__n_triangles

    @n_triangles.setter
    def n_triangles(self, n_triangles):
        if n_triangles > MAX_TRI:
            print('triangle overflow')
            n_triangles = MAX_TRI
            self.triangles = self.triangles[0:MAX_TRI-1]
            
        elif n_triangles < MIN_TRI:
            print('triangle underflow')
            diff = MIN_TRI - n_triangles
            for _ in range(diff):
                self.add_triangle
                self.n_triangles += 1


class DNA:
    channels = 3
    
    def __init__(self, resolution=225):
        self.resolution = resolution
        self.genes = Genes(self.resolution)
        
    def mutate(self, mutation_rate=0.01):
        rands = np.random.rand(self.genes.n_triangles)
        for rand, tri in zip(rands, self.genes.triangles):
            if rand < mutation_rate:
                tri.mutate(self.resolution)
                
        # adding new triangles
        if np.random.rand() < mutation_rate:
            self.genes.add_triangle()
        
    def crossover(self, partner):
        child = DNA()

        tri_space = self.genes.triangles + partner.genes.triangles
        tri_space_fit_dist = softmax([tri.fitness() for tri in tri_space])
        
        replace = False
        if child.genes.n_triangles >= len(tri_space):
            replace = True
            
        child.genes.triangles = np.random.choice(tri_space, size=child.genes.n_triangles,
                                                    replace=replace, p=tri_space_fit_dist).tolist()

#         idxs = np.argsort(tri_space_fit_dist)
    
#         temp = []
#         for idx in idxs:
#             temp.append(tri_space[idx])
            
#         child.genes.triangles = temp[0:child.genes.n_triangles]
            
        return child

    def fitness(self):
        pixels = self.get_rgb_repr()
        diff = sum(np.abs(pixels - target))
        target_tot = sum(target)

        return (target_tot - diff) / target_tot
        
    def get_rgb_repr(self):
        img_base = deepcopy(IMG_BASE)
        
        temp = []
        for tri in self.genes.triangles:
            cv2.fillPoly(img_base, tri.xyz, tri.rgb, cv2.LINE_AA)
            temp.append(deepcopy(img_base))
            
        dst = temp[0]
        for i in range(self.genes.n_triangles):
            if i == 0:
                pass
            else:
                alpha = 1.0/(i + 1)
                beta = 1.0 - alpha
                dst = cv2.addWeighted(temp[i], alpha, dst, beta, 0.0)
        
        return dst
        
    def show(self):
        plt.imshow(self.get_rgb_repr())
        plt.show()
```

```python
pop_size = 70
population = [DNA() for _ in range(pop_size)]
scores = []
```

```python
simulations = 100
mutation_rate = 0.05

start = time.time()

for i in range(simulations):
    for j in range(pop_size):
        
        BEST_FIT = population[np.argmax([inst.fitness() for inst in population])].get_rgb_repr()

        # fill pool from population
        pool = []
        pool_fitnesses = []
        fitnesses = [inst.fitness() for inst in population]
        
#         for inst, fitness in zip(population, fitnesses):
#             pool.append(inst)
#             pool_fitnesses.append(fitness)
        
        pool = population
        pool_fitnesses = fitnesses
    
        # get parents
        parent_a, parent_b = np.random.choice(pool, size=2, replace=False, p=softmax(pool_fitnesses))

        # crossover and mutate
        child = parent_a.crossover(parent_b)
        child.mutate(mutation_rate=mutation_rate)   

        #replace population member with new child
        population[j] = child

    score = sum([inst.fitness() for inst in population])/len(population)
    scores.append(score)
    print(i+1, round(score*100, 4), len(pool))
    
#     # show most fit inst
#     if(not (i+1) % 10): population[np.argmax([inst.fitness() for inst in population])].show()
    population[np.argmax([inst.fitness() for inst in population])].show()
        
    
print(time.time() - start)
```

```python
plt.plot(np.arange(len(scores)), scores)
plt.show()
```
