```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
import cv2
import time
from scipy.special import softmax
from copy import deepcopy
```

```python
target = cv2.imread("smiley.png")

class DNA:
    resolution = 225
    channels = 3
    
    def __init__(self, resolution=225):
        self.resolution = resolution
        self.genes = np.random.randint(0, 256, (self.resolution, self.resolution, self.channels), dtype=np.uint8)
        
    def mutate(self, mutation_rate=0.01):  
        mask_rands = np.random.rand(self.resolution, self.resolution, self.channels)
        mutation_rands = np.random.randint(0, 256, size=(self.resolution, self.resolution, self.channels))

        mutation_mask = mask_rands < mutation_rate
        mutation_mask_inv = mask_rands >= mutation_rate

        self.genes = np.multiply(self.genes, mutation_mask) + np.multiply(mutation_rands, mutation_mask_inv).astype(int)
        
    def crossover(self, partner):
        child = DNA()
                        
        rands = np.random.randint(2, size=(self.resolution, self.resolution, self.channels))
        rands_inv = np.ones((self.resolution, self.resolution, self.channels)) - rands
        child.genes = np.multiply(self.genes, rands) + np.multiply(partner.genes, rands_inv).astype(int)

        return child

    def fitness(self):
        genes_channeless = np.sum(self.genes, axis=2)
        target_channeless = np.sum(target, axis=2)
        return np.sum(self.genes == target[:,:,0:self.channels])**2 / (self.resolution**2*self.channels)**2
        #return np.sum(genes_channeless == target_channeless)**2 / self.resolution**4
        
    def show(self):
        plt.imshow(self.genes)
        plt.show()
```

```python
pop_size = 100
population = [DNA() for _ in range(pop_size)]
scores = []
```

```python
simulations = 200
mutation_rate = 0.0000001
#base = np.full((225,225,3), 255, dtype=np.uint8)
start = time.time()

for i in range(simulations):
    for j in range(pop_size):

        # fill pool from population
        pool = []
        pool_fitnesses = []
        fitnesses = [inst.fitness() for inst in population]
        
        pool = population
        pool_fitnesses = fitnesses
        
        # get parents
        parent_a, parent_b = np.random.choice(pool, size=2, replace=False, p=softmax(pool_fitnesses))

        # crossover and mutate
        child = parent_a.crossover(parent_b)
        child.mutate(mutation_rate=mutation_rate)   

        #replace population member with new child
        population[j] = child

    score = sum([inst.fitness() for inst in population])/len(population)
    scores.append(score)
    print(i+1, round(score*100, 4), len(pool))
    
    # show most fit inst
    if(not (i+1) % 10): population[np.argmax([inst.fitness() for inst in population])].show()
    
print(time.time() - start)
```

```python
plt.plot(np.arange(len(scores)), scores)
plt.show()
```
