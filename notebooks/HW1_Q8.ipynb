```python
import igraph as ig
```

```python
karate = ig.Graph.Read_GML('karate.gml')
ig.plot(karate)
```

```python
n_vertex = len(karate.vs)

inc_1s = karate.get_adjacency()[0]
inc_34s = karate.get_adjacency()[33]
colors = []
for inc_1, inc_34 in zip(inc_1s, inc_34s):
    if(inc_1 and inc_34):
        color = "orange"
    elif(inc_1):
        color = "red"
    elif(inc_34):
        color = "yellow"
    else:
        color = "blue"
    
    colors.append(color)
colors[0] = "red"
colors[33] = "yellow"

graph_style = {}
graph_style["vertex_size"] = [(v.degree()*2.5) for v in karate.vs]
graph_style["vertex_label"] = [i for i in range(0, n_vertex+1)]
graph_style["vertex_label_size"] = 15
graph_style["edge_label"] = [e.tuple for e in karate.es]
graph_style["edge_label_size"] = 7
graph_style["vertex_color"] = colors
```

This layout is used to apply an entire set of parameters to a network plotting method, such that the appropriate formatting, labeling, and other attribute settings are displayed. Igraph has numerous attributes available which enable pretty granular modification of edges, vertices, and graphs, making layouts a way to easily swap sets of attribute instructions.
