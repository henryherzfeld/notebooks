```r
library(neuralnet)
```

```r
train.labels <- sample(1:nrow(mtcars), .8*nrow(mtcars))
train.set <- mtcars[train.labels,]
train.set
test.set <- mtcars[-train.labels,]
```

```r
# Neuralnet package does not accept Label~. myForm replicates this behavior (using all features)
myform <- as.formula(paste('mtcars[,1] ~ ',paste(names(mtcars[!names(mtcars) %in% 'mtcars[,1]']), collapse = ' + ')))
```

```r
normalize <- function(data){
    vals <- NULL
    denom = max(data) - min(data)
    
    for(val in data){
        val <- (val - max(data))/denom
        vals <- c(vals, val)
    }
    return(matrix(vals))
}

mtcars$mpg <- normalize(mtcars$mpg)
```

```r
model <- neuralnet(mpg~cyl+disp+hp+drat+wt+qsec+vs+am+gear+carb, rep=20, data=train.set, hidden=2, linear.output = T)
model
```

```r
plot(model,rep="best")
```

```r
test.set
```

```r
predictions <- predict(model, test.set)
predictions
mean(test.set[,1])
```
