```python
import igraph as ig
import numpy as np
import matplotlib.pyplot as plt
```

```python
netscience = ig.Graph.Read_GML('netscience.gml')
```

Please report and draw the degree distributions of the whole network [0.25 pt].
Convert the figure to log-log space and validate whether it complies with the powerlaw distributions [0.25 pt]. 

```python
degree_dist = netscience.degree_distribution().bins()

xs = []
ys = []
for val in degree_dist:
    xs.append(val[0])
    ys.append(val[2])
```

```python
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.scatter(xs, ys, c='c', marker='x')
plt.show()

plt.xscale('log')
plt.yscale('log')
plt.xlim((1, max(xs)+5))
plt.ylim((0.9, max(ys)+100))
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.scatter(xs, ys, c='c', marker='x')
plt.show()
```

It appears that the line loosely fits the trend of a power-law distribution in a log-log space-- the line is roughly linear.

Please report cumulative degree distribution of the network [0.25 pt], and convert it
to log-log space and validate whether it complies with the power-law distributions
[0.25 pt].

```python
cum_ys = []
for i, x in enumerate(xs):
    cum_ys.append(sum(ys[i:]))
```

```python
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.scatter(xs, cum_ys, marker="x", c='c')
plt.show()

plt.xscale('log')
plt.yscale('log')
plt.xlim((1, max(xs)+5))
plt.ylim((0.9, max(cum_ys)+100))
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.scatter(xs, cum_ys, marker="x", c='c')
plt.show()
```

It appears that the line loosely fits the trend of a power-law distribution in a log-log space-- the line is roughly linear.

Please report Rank-Degree distribution of the network [0.25 pt], and convert it to
log-log space and validate whether it complies with the power-law distributions
[0.25 pt].

```python
degrees = [v.degree() for v in netscience.vs]
degrees_desc = sorted(degrees, reverse=True)
ranks = []
rank_degrees = []

for i, degree in enumerate(degrees_desc):
    ranks.append(i+1)
    rank_degrees.append(degree)
```

```python
plt.xlabel('Rank')
plt.ylabel('Degree')
plt.plot(ranks, rank_degrees)
plt.show()

plt.xscale('log')
plt.yscale('log')
plt.xlim((1, max(ranks)))
plt.ylim((0.9, max(rank_degrees)))
plt.plot(ranks, rank_degrees)
plt.xlabel('Rank')
plt.ylabel('Degree')
plt.show()
```

It appears that the line loosely fits the trend of a power-law distribution in a log-log space-- the line is roughly linear.
