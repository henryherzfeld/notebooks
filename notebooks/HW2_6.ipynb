Load Dataset

```r
housing <- read.table("http://www.cse.fau.edu/~xqzhu/courses/housing.header.binary.txt",header=T,sep=",")
attach(housing)
library("rpart")
```

Use “Crim” and “Rm” as independent variables to train an RPART decision tree

```r
fit <- rpart(Medv~Rm+Crim,method="class",data=housing)
```

Visualize the tree and explain the structure

```r
plot(fit, uniform="T")
text(fit,use.n=TRUE,all=TRUE,cex=.9)
```

This process leads to a leaf node. The leave nodes designate a classification as defined by their label. Labels are chosen based upon the purity threshold of a particular class presence through traversal of the decision nodes.
