To examine linear dependence between two sets of data, a scatterplot with each variable on each axis respectively might be sufficient. In this case, we can merely look at the distribution of data on the plane and examine their covariance. If a change in the first set's value corresponds to a change in the second set's value, we will examine a linear trend within the scatterplot, it's orientation dependant on the sign of the change in value. The density or sparsity of values around this trend indicates our standard deviation, where higher sparsity would reduce the correlation value. If there is no linear trend observed, it might be concluded the two sets of values are independent.
