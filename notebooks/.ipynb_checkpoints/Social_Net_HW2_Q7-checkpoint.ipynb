```python
import igraph as ig
from pandas import DataFrame
import numpy as np
from IPython.core.display import display, SVG
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
DataFrame(karate.shortest_paths())
```

```python
karate = ig.Graph.Read_GML('karate.gml')
```

```python
diameter = np.max(karate.shortest_paths())
print(f"The diameter of the network is {diameter}.")
```

```python
max_idxs = np.argwhere(karate.shortest_paths() == diameter)
```

```python
paths = []
for idx_arr in max_idxs:
    paths.append(karate.get_shortest_paths(idx_arr[0], to=idx_arr[1]))
```

```python
graph_style = {}

for path in paths:
    vertex_colors = ["red"]*len(karate.vs)
    
    for v in path[0]:    
        vertex_colors[v] = "green"
        
    graph_style["vertex_color"] = vertex_colors
    graph_style["bbox"] = (250, 250)
    plot_rep_svg, _ = ig.plot(karate, **graph_style)._repr_svg_()
    display(SVG(plot_rep_svg))
```

```python
print(list(map(round, karate.betweenness())))
print(list(map(round, karate.edge_betweenness())))
```

```python
max_v_betweenness = np.argmax(karate.betweenness())
max_e_betweenness = np.argmax(karate.edge_betweenness())

vertex_color = ["red"] * len(karate.vs)
vertex_color[max_v_betweenness] = "green"

edge_color = ["black"] * len(karate.es)
edge_color[max_e_betweenness] = "green"

graph_style = {}
graph_style["vertex_color"] = vertex_color
graph_style["edge_color"] = edge_color
graph_style["bbox"] = (400, 400)
ig.plot(karate, **graph_style)
```

```python
graph_style = {}
graph_style["bbox"] = (400, 400)
graph_style["vertex_size"] = [val*30 for val in karate.eigenvector_centrality()]

ig.plot(karate, **graph_style)
```
