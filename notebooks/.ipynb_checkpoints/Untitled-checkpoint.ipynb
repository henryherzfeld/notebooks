```r
# FOR 75% 25% RANDOM SPLIT
# set randomizer seed for sample function and sample housing dataset
set.seed(1)
train <- sample(1:nrow(housing),0.75*nrow(housing))
```

```r
# split dataset
train.set <- housing[train,]
test.set <- housing[-train,]
```

```r
# FOR K-FOLD CROSS VALIDATION
# shuffle housing dataset
housing.rand <- housing[sample(nrow(housing)),]
```

```r
# assign number of desired folds
fold.num <- 10
folds <- cut(seq(1,nrow(housing.rand)), breaks=fold.num, labels=FALSE)
# create dynamic array for containing accuracy
fold.acc <- c()
```

```r
# run classifier with data using kfold validation
for(i in 1:fold.num){
    # get indexes located at fold value
    testIndexes <- which(folds==i,arr.ind=TRUE)
    
    # make test and train using those indexes
    test <- housing.rand[testIndexes,]
    training <- housing.rand[-testIndexes,]
    
    # create classifier on training data
    fold.classifier <- rpart(Medv~.,method="class",data=training)
    
    # create predictions using test data
    fold.pred <- predict(fold.classifier, test)
    
    labels <- ifelse(fold.pred[,2]>.5,1,0)
    
    # assign accuracy to dynamic array
    fold.acc <- c(fold.acc, fold.pred)
    
    tt <- table(test$Medv, labels)
    
    acc <- ((tt[1][1] + tt[4][1])/(tt[1][1]) + tt[2][1] + tt[3][1] + tt[4][1])
    
    fold.acc <- c(fold.acc,acc)
}
```

```r
cv.mean.accuracy <- mean(fold.acc)
cv.mean.accuracy
```

```r
par(mfrow=c(5,2))
for (i in 1:fold.num){

    foldIndexes <- which(folds==i)
    
    
    plot(foldIndexes,fold.acc[foldIndexes])  
}
```
