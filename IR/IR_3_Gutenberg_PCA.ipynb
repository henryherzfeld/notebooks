# Word Embeddings Using FastText
Project Gutenberg provides free texts 

```python
import fasttext as ft
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

### Building Corpus

```python

cwd = os.getcwd()
books = []
books_dir = 'data/books'
books_path = os.path.join(cwd, books_dir)
for book in os.listdir(books_dir):
    path = os.path.join(cwd, books_dir, book)
    book_file = open(path, 'r')
    book_str = book_file.read()
    book_file.close()
    books.append(book_str)

corpus_str = "\n".join(books)
corpus_path = os.path.join(books_path, 'gutenberg_corpus.txt')
out = open(corpus_path, 'w')
out.write(corpus_str)
out.close()
```

```python
corpus_path = os.path.join(books_path, 'gutenberg_corpus.txt')
model = ft.train_unsupervised(corpus_path, model='skipgram')
```

```python
pca = PCA(n_components=2)
```

```python
word_vec_tups = [(word, model.get_word_vector(word)) for word in model.words]
word_vecs = [vec for _, vec in word_vec_tups]
```

```python
reduced = pca.fit_transform(word_vecs)
```

```python
start = 1300
end = 2000
offset = .001

plt.figure(figsize=(20,30))
for x, y, label in zip(reduced[start:end,0], reduced[start:end,1], model.words):
    plt.scatter(x, y)
    plt.text(x+offset, y+offset, label)

_ = plt.plot()
```
