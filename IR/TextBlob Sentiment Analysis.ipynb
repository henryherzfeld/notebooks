Import TextBlob

```python
from textblob import TextBlob
```

Read in Text Data

```python
with open('homework 3 data.txt') as file:
    text = file.read()
```

Split Sentences

```python
sentences = text.split('\t\n')
print(sentences[0])
```

Define polarity function, perform sentiment analysis

```python
def get_polarity(x):     
        if x > 0:
            return 'positive'
        elif x < 0:
            return 'negative'
        else: 
            return 'neutral'

for sentence in TextBlob(text).sentences[0:10]:
    print(sentence, get_polarity(sentence.sentiment.polarity), '\t', sep='\n')
```

Add CSV column to dataset

```python
with open('homework 3 data final.txt', 'w') as file:
    for sentence in TextBlob(text).sentences:
        file.write(f"{sentence}, {get_polarity(sentence.sentiment.polarity)}\n")
```
